## Misc functions to get data

```python
def get_velocities(r, times, sigma):
    for it,t in enumerate(times):
        Vi = r.GetVi(t, merged=True)
        print(type(Vi), dir(Vi), type(Vi['x']))
        Vix = Vi["x"][0]
        xV = Vi["x"][1][0]
        v = gaussian_filter1d(Vix(xV), sigma)
        if it==0 :
            vt = np.zeros((len(v),len(times)))
        vt[:,it] = v
    return xV, vt
```
